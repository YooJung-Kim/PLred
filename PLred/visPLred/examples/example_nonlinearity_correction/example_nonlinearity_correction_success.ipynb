```python
import PLred.visPLred.spec as sp
from scipy.sparse import load_npz
import json
import numpy as np
import PLred.visPLred.preprocess as pp
from astropy.io import fits
from tqdm import tqdm
import os
import matplotlib.pyplot as plt
```

```python
data = fits.getdata('/mnt/datazpool/PL/yjkim/reduced_map/betcmi_20250211/betcmi_20250211_couplingmap_frames_12:11:00_12:19:09.fits')
dark = fits.getdata('/mnt/datazpool/PL/yjkim/reduced/betcmi_20250211/dark.fits')
frame = data -dark[None,None,:,:]
```

```python
nonlinearity_model = '/mnt/datazpool/PL/yjkim/flat_characterization/2025-05-12/flux_dependent_flat/model'
model = pp.DetectorNonlinearityModel(modelname = nonlinearity_model)
```

```python
i,j = 3, 3

im_to_correct = frame[i,j,:, 1100:1300]
corrected_map, status = model.correct_map(im_to_correct)
```

```python
fig, axs = plt.subplots(ncols=2, figsize=(10,5))
axs[0].imshow(im_to_correct, vmin=0, vmax=500)
axs[0].set_title('Original Map')
axs[1].imshow(corrected_map, vmin=0, vmax=500)
axs[1].set_title('Corrected Map')

for ax in axs:
    ax.set_xlim([80,130])
    ax.set_ylim([60,90])
plt.tight_layout()
```
